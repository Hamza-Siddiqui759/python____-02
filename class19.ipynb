```python
# write the function is_anagram
def is_anagram(test, original):
    test = sorted(test.lower())
    original = sorted(original.lower())
    
    return test == original
    
```

```python
is_anagram('efofet','toffe')
```

```python
def remove_duplicate(s):
    seen = set()
    result = ""
    for char in s:
        if char.lower() not in seen:
            result += char
            seen.add(char)
    return result

```

```python
remove_duplicate('hello')
```

```python
def find_vowels(s):
    reuslt = ("")
    for vowel in s:
        if vowel not in 'aeiou':
            reuslt += vowel
    return reuslt
```

```python
find_vowels('hello world')
```

```python
def longest_word(ss):
    
    words = ss.split()
    longest = ''
    for word in words:
        if len(word) > len(longest):
            longest = word
    return longest
```

```python
longest_word('i love python')
```

```python
def duplicate_words(s):
    sent = set()
    duplicate = ''
    for catch_duplicate in s:
        if catch_duplicate not in sent:
            duplicate += catch_duplicate
            sent.add(catch_duplicate)
        
    return duplicate
```

```python
duplicate_words('hello')
```
