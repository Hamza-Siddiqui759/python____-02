```python
class Resturant():
    def __init__(self,resturant,menu):
        self.resturant = resturant
        self.menu = menu

    def resturant_name(self):
        print(f'{self.resturant}------{self.menu}')

    def meno(self):
        print(f"This {self.resturant} offers {self.menu}")
            
```

```python
classs = Resturant('Kababjees','zinger')
```

```python
classs.resturant_name()
```

```python
classs.meno()
```

```python
class Cuisine():
    def __init__(self,resturant,cuisine):
       self.resturant = resturant
       self.cuisine = cuisine
       self.make  = f"{resturant} {cuisine}"
    
    def open(self):
        print(f'This resturant neme called {self.resturant} \nserves {self.cuisine}')


    def close(self):
        print(f'This resturant {self.make} is closed today')
```

Modifying Attribute Values

```python
class Using():
    def __init__(self, first_name, last_name, gender, age):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender 
        self.age = age 
        self.campus = 'JDC'
    def calss(self):
       print( f' Users name : {self.first_name} {self.last_name}\n Gender : {self.gender}\n Age : {self.age}\n Campus : {self.campus}')
    
    def calling(self):
        print(f'you credentials are correct {self.first_name}.')

    def change(self,new_campus = 'SMIT'):
        self.campus = new_campus   
```

```python
call = Using('Ali','Khan','Male',24)
kashan = Using('Kashan', 'khan', "Male", 20)
yaman = Using('Yaman', 'peshawari', "Male", 10)
ghafar = Using('Ghafar', 'sawati', "Male", 12)
```

```python
#Modifying an Attributes Value Directly

kashan.campus = "JDC"
```

```python
kashan.campus
```

```python
yaman.campus = 'JDC'
```

```python
yaman.campus
```

```python
call.calss()
kashan.calss()
yaman.calss()
ghafar.calss()
```
