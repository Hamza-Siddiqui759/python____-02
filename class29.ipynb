Base Class (ElectronicDevice): This is the parent class: Attribute: name, price, power_status
Methods: (power_on()) that all electronic devices share.

Subclass (Laptop): This is the child class. It inherits (automatically receives) all the attributes and methods from ElectronicDevice. Attributes:(screen_size_inches, ports) Methods: list_ports()


```python
class Car():
    def __init__(self,name , price,power_status):

        self.name = name
        self.price = price
        self.power_status = power_status
    def power_on(self):
        print('power is on')
```

```python
class ELECTRONIC(Car):
    def __init__(self, name, price, power_status,screen_size,ports):
        super().__init__(name, price, power_status)
        self.screen_size = screen_size
        self.ports = ports
    def list_ports(self):
        print(f'These are the ports {' , '.join(self.ports)}')
```

```python
d1 = ELECTRONIC('Ali', 2000 ,'ON',20.2,['hdmi'  , 'vg'])
```

```python
d1.list_ports()
```

Instances as Attributes

```python
class Car():
    def __init__(self,make,model,year):
        self.make = make
        self.model = model 
        self.year = year

    def apply_acclerator(self):
        print(f'Apply acceleratir {self.model}')
    def apply_brakes(self):
        print(f'Apply brakes {self.make} {self.model}')
```

```python
class Battery():
    def __init__(self,company,capacity,voltage):
        self.company = company
        self.voltage = voltage
        self.capacity = capacity
    def ook(self):
        print(f'THE CAR HAS BEST {self.company} AND {self.voltage}, {self.capacity}')
```

```python
class ELECTRONIC(Car):
    def __init__(self, make, model, year,company,capacity,voltage):
        super().__init__(make, model, year)

        self.batteryy = Battery(company,capacity,voltage)

    def battery(self):
        print(f'{self.batteryy} is charges now')
```

```python
newec = ELECTRONIC('Honda', 'Civic', 2025,'osaka','12V', '55AH')
```

```python
newec.batteryy.ook()
newec.batteryy.company
```

```python
newec.apply_acclerator()
```

```python
newec.apply_brakes()
```

```python
class ElectricCar(Car):
    def __init__(self,make, model, year, battery):
        super().__init__(make,model,year)

        # Additional Attribute
        self.battery = battery

    # Additional Method
    def recharge(self):
        print('The battery is recharged now ')
```

```python
b2 = Battery('AGs', '150v', '200AH')
```

```python
ec3 = ElectricCar('Honda', 'Civic', 2021, b2)
```

```python
ec3.battery.ook()
```
