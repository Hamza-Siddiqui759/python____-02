```python
# print those city/cities having all characters unique (not repetitive)
cities = ['Karachi', 'Lahore', 'Quetta', 'Dadu', 'Murre']
```

```python
for city in cities:
    count = True
    for char in city:
        if city.lower().count(char) > 1:
            count = False
              
    # if charcount == True:
    if count:
        print(city)
```

```python
for city in cities:
    empty_set = set()    
    for char in city.lower():
        empty_set.add(char)
    if len(empty_set) == len(city):
        print(city)
```
