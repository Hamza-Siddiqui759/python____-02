Binding of methods with attributes

```python
class User:
    def __init__(self, first_name, last_name, gender, age, campus = "SMIT"):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = campus

    def change_campus(self, new_campus):
        self.campus = new_campus

    def show_campus(self):
        return self.campus

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')
```

```python
car = User('ali','khan','male',24)
```

```python
car.change_campus('sir sahil') #SETTER
```

```python
car.show_campus()     #GETTER
```

```python
class IN():
    def __init__(self,fname,lname,age,gender,campus = 'SMIT'):
       self.fname = fname
       self.lname = lname
       self.age =age
       self.gender = gender
       self.campus = campus

    def setter(self,new_fname = '',new_lname ='',new_age = '',new_gender = '',new_campus = ''):
        if self.fname:
         self.fname = new_fname
        if self.lname:
         self.lname = new_lname
        if self.age:
           self.age = new_age
        if self.gender:
         self.gender = new_gender
        if self.campus:
         self.campus = new_campus

    def getter(self):
       print(self.fname)
       print(self.lname)
       print(self.age)
       print(self.gender)
       print(self.campus)
    def describe_user(self):
        print(f"User's name : {self.fname} {self.lname}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")
```

```python
get = IN('ali','khan',24,'male','smit')
```

```python
get.setter('faheem','khan')
```

```python
get.getter()
```

```python
get.describe_user()
```

```python
get.__dict__
```

```python
class USERS():
    def __init__(self,fname,lname,age,gender,campus = 'SMIT'):
       #Attribute
       self.fname = fname
       self.lname = lname
       self.age =age
       self.gender = gender
       self.campus = campus
       self.counter = 1
       self.getattribute()
       #METHODS
    def getattribute(self):
       data = self.__dict__
       with open('testing.txt', 'a+')as f:
           f.write(f'Users {self.counter}\n')
           for attr,value in data.items():
               self.counter += 1
               if '-' in attr:
                   ''.join(attr).split(', ')
               else:
                   attr.title()
                   f.write(f'{attr} : {value} \n')  
           f.write('\n')                           
               
```

```python
use = USERS('ali','khan',24,'male','smit')
```
