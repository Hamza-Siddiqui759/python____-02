Access Modifiers / Specifiers

```python
class Modifiers():
    def __init__(self,fname,lname,gender,age,campus = 'SMIT'):
        self.fname = fname
        self.lname = lname 
        self.gender = gender
        self._age = age
        self.__campus = campus    #private attribute we cant access it a directly and also cant change it directly we had to use setter getter
        #for this .the benefit of private attributer that no one can change your attribute directly the user had to change by using methods.
    def setter(self,new_campus):
        self.__campus = new_campus

    def getter(self):
        return self.__campus
    
    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")
    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')
        
        
```

```python
classs = Modifiers('ali','khan','male',24)
```

```python
classs.setter('jdc')
```

```python
classs.getter()
```
