```python
class User:
    def __init__(self, first_name, last_name, gender, age):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = 'SMIT'

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')
    
    def setattribute(self,attribute,value):
        setattr(self,attribute,value)
```

```python
temp = User('ali', 'khan', 'male', 23)
```

```python
temp.setattribute('first_name', 'kashan')
```

```python
class Car:
    def __init__(self,make, model, year):
        # Attributes
        self.make = make
        self.model = model
        self.year =  year 
    
    # Methods
    def apply_accelerator(self):
        print('Accelerator is applied')
    
    def apply_breaks(self):
        print('Brakes are applied')
    
```

```python
class ElectricCar():
    def __init__(self,make, model, year, battery):
        # Attributes
        self.make = make
        self.model = model
        self.year =  year 
        # Additional Attribute
        self.battery = battery
    
    # Methods
    def apply_accelerator(self):
        print('Accelerator is applied')
    
    def apply_breaks(self):
        print('Brakes are applied')

    # Additional Method
    def recharge(self):
        print('The battery is recharged now ')
    
```

```python
ec1 = ElectricCar('toyota', 'corrola', '2025', 'Osaka')
```

```python
ec1.recharge()
```

```python
class Car:
    def __init__(self,make, model, year):
        # Attributes
        self.make = make
        self.model = model
        self.year =  year 
    
    # Methods
    def apply_accelerator(self):
        print(f'Accelerator is {self.model} applied')
    
    def apply_breaks(self):
        print(f'Brakes are applied {self.model}')
```

```python
class ElectricCar(Car):
    def __init__(self, make, model, year, battery):
        super().__init__(make, model, year)     #BASICALLY SUPER IS USED TO PULL CHILD CLASS METHODS AND ATTRIBUTES
     #   additional attribute
        self.battery = battery

     #Additional methods
    def charge(self):
        print(f'the battery {self.battery} is charged now')
```

```python
d = ElectricCar('honda','civic' , 2022 , 'osaka' )
```

```python
d.charge()
```

```python
d.apply_accelerator()
```
