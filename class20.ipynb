```python
def water(time):
  water = 0.5
  if time:
      wites = water*time
      return int(wites)
      
```

```python
water(11.8)
```

```python
def digitize(n):
    result = []
    for num in str(n)[::-1]:
        result.append(int(num))
    return result
```

```python
digitize(21313)
```

```python
def functions(word,letter):
   if word:
     wipes = word.count(letter)
     return wipes
   else:
      return 0

            
```

```python
def change_count(change):
    data = {
        'penny': 0.01,
        'nickel': 0.05,
        'dime': 0.10,
        'quarter': 0.25,
        'dollar': 1.00,
    }
    total = 0
    for word in change.split(' '):
        
        total += data.get(word)
    result = f"${total:.2f}"
    return result
```

```python
change_count('nickel')
```
